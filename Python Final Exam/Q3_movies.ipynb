```python
import pandas as pd

# Load the dataset
file_path = 'movies.csv'
movies_df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
movies_df.head()
```

```python
movies_df.describe()
```

```python
grp_movies = movies_df.groupby("Genre") 

Genre_Count_list = []

for grp, grp_info in grp_movies:
    Genre_Count_dict = {}
    Genre_Count_dict["Unique_Movie_Genres"] = grp

    grp_res = grp_movies.get_group(grp)
    
    for index, row in grp_res.iterrows():
        if row ["Genre"] not in Genre_Count_dict.keys():
            Genre_Count_dict[row["Genre"]] = [row["Film"]]
        else:
            if row["Film"] not in Genre_Count_dict.values():
                Genre_Count_dict[row["Genre"]].append(row["Film"])
    Genre_Count_list.append(Genre_Count_dict)

Genre_Count_list
```

```python
for row in Genre_Count_list:
    for key, value in row.items():
        item = {}
        if isinstance(value, list):
            for v in value:
                if v not in item.keys():
                    item[v] = 1
                else:
                    item[v] += 1
            row[key] = item

Genre_Count_list
```

```python
grp_movies = movies_df.groupby("Genre") 

Audience_Score_list = []

for grp, grp_info in grp_movies:
    Audience_Score_dict = {}
    Audience_Score_dict["Unique_Movie_Genres"] = grp

    grp_res = grp_movies.get_group(grp)
    
    for index, row in grp_res.iterrows():
        if row ["Genre"] not in Audience_Score_dict.keys():
            Audience_Score_dict[row["Genre"]] = [row["Audience score %"]]
        else:
            if row["Audience score %"] not in Audience_Score_dict.values():
                Audience_Score_dict[row["Genre"]].append(row["Audience score %"])
    Audience_Score_list.append(Audience_Score_dict)

Audience_Score_list
```

```python
for row in Audience_Score_list:
    for key, value in row.items():
        item = {}
        if isinstance(value, list):
            item["Min"] = min(value)
            item["Mean"] = round((sum(value) / len(value)), 2)
            item["Max"] = max(value)
            row[key] = item

Audience_Score_list
```
